# Статистичні величини

```python
vote_counts = np.array(df_movies.vote_count)
vote_counts
```

### 1. Вибіркове середнє значення

```python
def sample_average(sample):
    average = 0

    for item in sample:
        average += item

    return average / sample.size

print(sample_average(vote_counts))
print('numpy:', vote_counts.mean())
```

### 2. Вибіркова дисперсія

```python
def sample_variance(sample):
    average = sample_average(sample)
    variance = 0

    for item in sample:
        variance += np.square(item - average)

    return variance / sample.size

print(sample_variance(vote_counts))
print('numpy:', vote_counts.var())
```

### 3. Стандартне відхилення

```python
def standard_deviation(sample):
    variance = sample_variance(sample)

    return np.sqrt(variance)

print(standard_deviation(vote_counts))
print('numpy:', vote_counts.std())
```

### 4. Медіана

```python
def median(sample):
    sort_sample = np.sort(sample)
    size = sort_sample.size

    if size % 2 == 0:
        return (sort_sample[size / 2] + sort_sample[(size / 2) - 1]) / 2

    return sort_sample[size // 2]

print(median(vote_counts))
print('numpy:', np.median(vote_counts))
```

### 5. Мода

```python
def sample_mode(sample):
    unique, counts = np.unique(sample, return_counts=True)
    amount_of_elements = dict(zip(unique, counts))

    key = sample[0]
    results = dict({key: amount_of_elements[key]})

    for keys in amount_of_elements:
        if amount_of_elements[keys] > results[key]:
            results = dict({keys: amount_of_elements[keys]})
            key = keys
        elif amount_of_elements[keys] == results[key]:
            results[keys] = amount_of_elements[keys]

    return results

mode = sample_mode(vote_counts)
for k in mode:
    print('Елемент: ' + '"' + str(k) + '"' + ', кількість:', mode[k])
```

### 6. Максимальне і мінімальне значення

```python
def sample_max(sample):
    result = sample[0]

    for item in sample:
        if item > result:
            result = item

    return result

def sample_min(sample):
    result = sample[0]

    for item in sample:
        if item < result:
            result = item

    return result

print('Мінімальне значення:', sample_min(vote_counts))
print('Максимальне значення:', sample_max(vote_counts))
```

### 7. Розмах

```python
def sample_span(sample):
    return sample_max(sample) - sample_min(sample)

print(sample_span(vote_counts))
```

### 8. Квантилі: 0.1, 0.25, 0.5, 0.75

```python
def quantile(sample, q):
    sort_sample = np.sort(sample)

    return sort_sample[int(sample.size * q)]

print('0.1: ', quantile(vote_counts, 0.1))
print('0.25:', quantile(vote_counts, 0.25))
print('0.5: ', quantile(vote_counts, 0.5))
print('0.75:', quantile(vote_counts, 0.75))
print()
print('numpy: ')
print('0.1: ', np.quantile(vote_counts, 0.1))
print('0.25:', np.quantile(vote_counts, 0.25))
print('0.5: ', np.quantile(vote_counts, 0.5))
print('0.75:', np.quantile(vote_counts, 0.75))
```

# Метрика популярності фільмів

### Рейтинг

```python
def weighted_rating(data_frame, c_is_median = False):
    v = data_frame.vote_count # кількість оцінок за фільм
    m = data_frame.vote_count.min() # мінімальна кількість оцінок у вибірці
    r = data_frame.vote_average # середній рейтинг конкретного фільму
    c = data_frame.vote_average.median() if c_is_median else data_frame.vote_average.mean()

    result = (v * r / (v+m)) + (m*c/(v+m)) # рейтинг фільмів

    return result
```

### Top-10 фільмів у тренді

```python
def top_movies(data_frame, wr, amount, color, figsize, genre='фільмів'):
    wr = wr.sort_values()[wr == wr]

    plt.figure(figsize=figsize)
    plt.barh(data_frame.title_x[wr.index].tail(amount), wr.tail(amount), align='center', color=color)
    plt.xlabel("Рейтинг", fontsize=18)
    plt.title("Top-" + str(amount) + " " + genre + " у тренді", fontsize=24)
    plt.show()

WeightedRating = weighted_rating(df)
top_movies(df, WeightedRating, 10, 'm', (15,10))
```

### Top-5 фільмів для різних жанрів

```python
def select_genres_id(films, genre):
    results = []

    for series in films.iterrows():
        genres = json.loads(series[1].genres)

        for item in genres:
            if item['name'] == genre:
                results.append(series[1].id)

    return results
```

```python
def top_five_actions(data_frame, c_is_median=False):
    ids_action = select_genres_id(data_frame[['genres', 'title_x', 'id']], "Action")
    films_action = data_frame.loc[data_frame['id'].isin(ids_action)]
    weighted_rating_action = weighted_rating(films_action, c_is_median)

    top_movies(films_action, weighted_rating_action, 5, 'b', (10, 5), 'Action')
```

```python
top_five_actions(df)
```

```python
def top_five_fantasy(data_frame, c_is_median=False):
    ids_fantasy = select_genres_id(data_frame[['genres', 'title_x', 'id']], "Fantasy")
    films_fantasy = data_frame.loc[data_frame['id'].isin(ids_fantasy)]
    weighted_rating_fantasy = weighted_rating(films_fantasy, c_is_median)
    top_movies(films_fantasy, weighted_rating_fantasy, 5, 'c', (10, 5), 'Fantasy')
```

```python
top_five_fantasy(df)
```
